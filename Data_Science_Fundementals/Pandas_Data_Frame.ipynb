```python
# import required packages
import numpy as np
import pandas as pd
```

```python
# explicitly create dataframe form np.array

# initialize array
x = np.array([[0,1,2,3],[4,5,6,7]])

# explicitly create dataframe
df = pd.DataFrame(x)

#out
df
```

```python
# renaming of rows and columns

# rename rows (indecies)
df.index = ['row_1','row_2']

# rename columns
df.columns = ['first','this', 'then', 'that']

#out
df
```

```python
# setting of row and col names on DataFrame initialization
df2 = pd.DataFrame(
                    x,
                    columns=['first','this','then','that'],
                    index=['row_1','row_2']
                    )
    
#out
df2
```

```python
# Starting with an empty data frame with rows named from list

# This list will become our row names.
names = ['George',
         'John',
         'Thomas',
         'James',
         'Andrew',
         'Martin',
         'William',
         'Zachary',
         'Millard',
         'Franklin']

# Create an empty data frame with named rows.
purchases = pd.DataFrame(index=names)

# Add our columns to the data frame one at a time.
purchases['country'] = ['US', 'CAN', 'CAN', 'US', 'CAN', 'US', 'US', 'US', 'CAN', 'US']
purchases['ad_views'] = [16, 42, 32, 13, 63, 19, 65, 23, 16, 77]
purchases['items_purchased'] = [2, 1, 0, 8, 0, 5, 7, 3, 0, 5]
purchases 
```

```python
# generating new column based on two existing columns
purchases['items_purch_per_ad'] = purchases['items_purchased'] / purchases['ad_views']
purchases
```

```python
# also possible in dot notation
purchases9ads_per_purch = purchases.ad_views/purchases.items_purchased
purchases
```
