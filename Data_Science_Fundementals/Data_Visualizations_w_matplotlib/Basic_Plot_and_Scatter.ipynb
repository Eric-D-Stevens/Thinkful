```python
# generate random dataset
# Make the random function consistent and replicable.
np.random.seed(1221)

# Make a blank data frame.
df = pd.DataFrame()

# Add a column of random numbers between 0 and 1.
df['rand'] = np.random.rand(100) # 100 random numbers
df['rand_sq'] = df['rand'] ** 2 # numbers in 'rand' squared
df['rand_shift'] = df['rand'] + 2 # numbers in 'rand' + 2

# When creating a data frame an index column of counts is created, counting from 0.
# Here we do a few transforms on that index to create some extra columns.
df['counts_sq'] = df.index ** 2 
df['counts_sqrt'] = np.sqrt(df.index)
```

```python
# display the rand graph
%config InlineBackend.figure_format = 'svg'
plt.plot(df['rand'])
plt.show()
```

```python
# changing plot options

# plot color purple
plt.plot(df['rand'], color='purple')

# set height of graph
plt.ylim([-0.5,1.5])

# set y axis label
plt.ylabel('The Values')

# set graph title
plt.title('Random Stuff')

# display
plt.show()
```

```python
# adding multiple lines to single graph

# plot random red
plt.plot(df['rand'], color='red')

# plot shift orange
plt.plot(df['rand_shift'], color='orange')

# set y range
plt.ylim([-0.5,3.5])

# set y label
plt.ylabel('The Values')

# set graph title
plt.title('Two Random Stuffs')
```

```python
# Plotting two columns agains eachother with .scatter()

# declare the x and y
plt.scatter(x=df['rand'], y=df['rand_sq'])

# display
%config InlineBackend.figure_format = 'svg'
plt.show()
```

```python
# adding option to scatter plots

# plot rand agains rand_sq in green
plt.scatter(
            x=df['rand'],
            y=df['rand_sq'],
            color='green',
            marker='o', s=11
            )
            
# plot rand agains rand_shift in yellow 
plt.scatter(
            x=df['rand'],
            y=df['rand_shift'],
            color='yellow',
            marker='o', s=11
            )

# show
%config InlineBackend.figure_format = 'svg'
plt.show()
```

```python
# built in panda power plots

# scatter plot declaring x and y
df.plot(kind='scatter', x='counts_sq',y= 'counts_sqrt')

# plot all in line
df.plot(kind='line')

# counts_sq overpowering, remove
dfx = df.drop(['counts_sq'], axis=1).plot(kind='line')

# show all plots
plt.show()

```
